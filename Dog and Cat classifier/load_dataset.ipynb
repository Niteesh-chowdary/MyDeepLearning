```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
DATADIRS = ["E:\\deep learning data\\image recognition\\training_set",
            "E:\\deep learning data\\image recognition\\test_set"] #mention the paths of your train and test data
CATEGORIES = ['dogs','cats']
for datadir in DATADIRS:
    for category in CATEGORIES:
        path = os.path.join(datadir,category)
        for img in os.listdir(path):
            img_array = cv2.imread(os.path.join(path,img))
            plt.imshow(img_array)
            plt.show()
            break
        break
```

```python
IMG_SIZE = 64
new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
plt.imshow(new_array)
plt.show()
```

```python
train_data = []


def create_training_data():
    for datadir in DATADIRS:
        for category in CATEGORIES:
            path = os.path.join(datadir,category)
            class_num = CATEGORIES.index(category)
            for img in os.listdir(path):
                try:
                    img_array = cv2.imread(os.path.join(path,img))
                    new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                    train_data.append([new_array,class_num])
                except Exception as e:
                    pass

create_training_data()
```

```python
print(len(train_data))
```

```python
import random
```

```python
random.shuffle(train_data)
```

```python
for sample in train_data:
    print(sample[1])
```

```python
X = []
Y = []

for features, label in train_data:
    X.append(features)
    Y.append(label)
    
X = np.array(X).reshape(-1,IMG_SIZE,IMG_SIZE,3)
Y = np.array(Y).reshape(-1,1)
```

```python
print(Y.shape)
```

```python
# storing the data set
import pickle 

pickle_out = open("X_traincat.pickle",'wb')
pickle.dump(X,pickle_out)
pickle_out.close()

pickle_out = open("Y_traincat.pickle",'wb')
pickle.dump(Y,pickle_out)
pickle_out.close()
```
